```python
import pandas as pd

new_songs = pd.read_csv('30k songs data\spotify_songs.csv')

new_songs
```

```python
#songs included in user_profiling

user_profile_data = pd.read_csv('user_profile_data.csv')

user_profile_data
```

```python
user_profile_data.columns

user_profile_data = user_profile_data.drop(['Artist ID', 'time_signature', 'Genre 2', 'Genre 3'], axis=1)
```

```python
new_songs.columns
```

```python
song_columns_to_drop = ['track_id', 'track_popularity',
       'track_album_id', 'track_album_name', 'track_album_release_date',
       'playlist_name', 'playlist_id', 'playlist_subgenre']

new_songs = new_songs.drop(song_columns_to_drop, axis=1)

new_songs
```

```python
new_songs.columns
```

```python
# reformat the column names to match the user_profile_data

new_songs.columns = ['track', 'artist', 'genre', 'danceability',
       'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness',
       'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms']

new_songs
```

```python
new_songs.columns

# add in userid column

new_songs['userid'] = 0
```

```python
new_songs
```

```python
new_songs.columns
```

```python
user_profile_data.columns
```

```python
# renamed user_profile_data columns to match new_songs columns

user_profile_data.columns = ['artist', 'userid', 'track', 'playlist', 'danceability', 'energy',
       'key', 'loudness', 'mode', 'speechiness', 'acousticness',
       'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms',
       'genre']

user_profile_data
```

```python
# reorder the columns to match new_songs

user_profile_data = user_profile_data[['track', 'artist', 'genre', 'danceability', 'energy', 'key', 'loudness',
       'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness',
       'valence', 'tempo', 'duration_ms', 'userid']]

user_profile_data.columns
```

```python
new_songs.columns
```

```python
# merge the two dataframes

merged_data = pd.concat([user_profile_data, new_songs])

merged_data
```

```python
merged_data.to_csv('merged_data_allens_request.csv', index=False)
```
