# Final Data Merging

importing the genres associated with artist

```python
import pandas as pd

genre_inital_df = pd.read_csv('All_Generated_Genres_with artists\spotify_data_nick_generated.csv')

genre_inital_df
```

```python
genre_inital_df.columns
```

```python
# drop everything except the genre, artist, artist id columns

genre_df = genre_inital_df.drop(columns=['Unnamed: 0', 'danceability', 'energy', 'key', 'loudness', 'mode',
       'speechiness', 'acousticness', 'instrumentalness', 'liveness',
       'valence', 'tempo', 'durationms', 'timesignature'])

genre_df
```

load the audio features list

```python
generated_audio_features = pd.read_csv('Spotify-User-Playlist-Audio-features\gen_audio_features.csv', on_bad_lines='skip')

generated_audio_features
```

```python
# make new column headers, move the current column headers to the first row, and reset the index

column_names = ['track', 'danceability', "energy","key", "loudness", "mode",
                    "speechiness", "acousticness", "instrumentalness", "liveness", "valence", "tempo", "duration_ms", "time_signature"]

#move current column names to the first row
generated_audio_features = generated_audio_features.T.reset_index().T.reset_index(drop=True)
generated_audio_features.columns = column_names 

generated_audio_features
```

add in the spotify user database so we can make a key to match everything together

```python
playlist_data = pd.read_csv('Spotify-User-Only-Playlist-Data\spotify_dataset.csv', on_bad_lines='skip')

playlist_data
```

```python
playlist_data.columns
```

```python
playlist_data.columns = playlist_data.columns.str.replace('"', '').str.replace(' ', '_').str.lower()

playlist_data.columns = playlist_data.columns.str.replace('name', '')

playlist_data.columns = playlist_data.columns.str.replace('_', '')

playlist_data.columns
```

```python
playlist_data
```

```python
# merge the genre_df and playlist_data dataframes on the artist column

merged_df = pd.merge(genre_df, playlist_data, on='artist')

merged_df
```

```python
# merge the merged_df and generated_audio_features dataframes on the track column

merged_with_features = pd.merge(merged_df, generated_audio_features, on='track')

merged_with_features
```

```python
# keep artists that have a genre and audio features

merged_with_features = merged_with_features.dropna()

merged_with_features
```

```python
if 'Genre' in merged_with_features.columns:
    genre_expanded = merged_with_features['Genre'].str.split(',', expand=True)
    
    for i in range(genre_expanded.shape[1]):
        genre_expanded.rename(columns={i: f'Genre {i+1}'}, inplace=True)
    merged_genre_split_features = pd.concat([merged_with_features, genre_expanded], axis=1)
    merged_genre_split_features.drop(columns='Genre', axis=1, inplace=True)
else:
    print('No genre column found')
    
merged_genre_split_features
```

```python
# count of "None" in each column

merged_genre_split_features.replace('None', pd.NA, inplace=True)
merged_genre_split_features.isna().sum()
```

```python
# drop all genre columns except 1, 2, 3

merged_genre_split_features.drop(columns=['Genre 4', 'Genre 5', 'Genre 6', 'Genre 7', 'Genre 8', 'Genre 9', 'Genre 10', 'Genre 11', 'Genre 12', 'Genre 13', 'Genre 14'], inplace=True)

merged_genre_split_features
```

```python
merged_genre_split_features.to_csv('user_profile_data.csv', index=False)
```

```python
# save a version of this dataframe that is about half the size

merged_genre_split_features_half = merged_genre_split_features.sample(frac=0.5)

merged_genre_split_features_half.to_csv('user_profile_data_half.csv', index=False)
```
