```python
import pandas as pd
audio_features = pd.read_csv('audio_features.csv', on_bad_lines='skip')

audio_features
```

```python
column_names = ['track', 'danceability', "energy","key", "loudness", "mode",
                    "speechiness", "acousticness", "instrumentalness", "liveness", "valence", "tempo", "duration_ms", "time_signature"]

#move current column names to the first row
audio_features = audio_features.T.reset_index().T.reset_index(drop=True)
audio_features.columns = column_names 

audio_features
```

```python
user_data = pd.read_csv("Spotify-User-Only-Playlist-Data\spotify_dataset.csv", on_bad_lines='skip')

user_data
```

```python
#merge the two dataframes on the the 

#match wherever the artist name in the playlist data matches the artist name in the spotify data
user_data.columns

#rename all column names to exclude quotes "" and spaces
user_data.columns = user_data.columns.str.replace('"', '').str.replace(' ', '_').str.lower()
#rename all column names to exclude _
user_data.columns = user_data.columns.str.replace('_', '')
#drop 'name' in each column
user_data.columns = user_data.columns.str.replace('name', '')

user_data.columns
```

```python
# Strip whitespace from the 'track' column in both datasets
audio_features['track'] = audio_features['track'].str.strip()
user_data['track'] = user_data['track'].str.strip()

# merge dataset on the track column, only include rows where the track name is the same in both datasets, drop everything else
merged_data = pd.merge(audio_features, user_data)
```

```python
merged_data
```

```python
# drop all occurences of "Light Speed" track name except the first one

merged_data = merged_data.drop_duplicates(subset='track', keep='first')
```

```python
merged_data
```

```python
# reset index
merged_data = merged_data.reset_index(drop=True)
```

```python
merged_data
```

```python
# Calculate the midpoint of the DataFrame
midpoint = len(merged_data) // 2

# First half of the DataFrame
first_half = merged_data.iloc[:midpoint]

# Second half of the DataFrame
second_half = merged_data.iloc[midpoint:]

# Print both halves
print("First Half:")
print(first_half)
print("\nSecond Half:")
print(second_half)
```
