```python
# Define function to calculate mode of a list
def calculate_mode(lst):
    if len(lst) == 0:
        return None
    counter = Counter(lst)
    max_count = max(counter.values())
    modes = [k for k, v in counter.items() if v == max_count]
    if len(modes) == 1:
        return modes[0]
    else:
        # If there are multiple modes, return one of them arbitrarily
        return modes[0]
    
# Define function to calculate average of a list
def calculate_average(lst):
    if len(lst) == 0:
        return None
    return sum(lst) / len(lst)

# Initialize variables to keep track of previous user ID and playlist
prev_user = None
prev_playlist = None

# Initialize an empty list to store user profiles
user_profiles = []

# Iterate over each row in the DataFrame
for index, row in df.iterrows():
    # If new user ID or new playlist, append profile for previous user and reset output dictionary
    if row['userid'] != prev_user or row['playlist'] != prev_playlist:
        if prev_user is not None and prev_playlist is not None:
            user_profiles.append(output)
        output = {}  # Reset output dictionary
        prev_user = row['userid']
        prev_playlist = row['playlist']

    # Drop userid and playlist columns
    data = row.drop(labels=['userid', 'playlist'])

    # Calculate mode for string columns and average for numerical columns
    for col in data.index:
        if col.startswith('Genre'):
            output[col] = calculate_mode([data[col]])
        else:
            if col in output:
                if isinstance(output[col], int) or isinstance(output[col], float):
                    output[col] = (output[col] + data[col]) / 2
            else:
                output[col] = data[col]

# Append the last user profile to the list
user_profiles.append(output)

# Convert the list of dictionaries to a DataFrame
profile_df_nick = pd.DataFrame(user_profiles)

profile_df_nick
```

```python
import pandas as pd
from collections import Counter

# Define function to calculate mode of a list
def calculate_mode(lst):
    if len(lst) == 0:
        return None
    counter = Counter(lst)
    max_count = max(counter.values())
    return [k for k, v in counter.items() if v == max_count][0]

# Group by 'userid' and aggregate using custom functions for each column
aggregation_functions = {
    'userid': 'first',  # Just to ensure userid is retained in the final output
    'artist': lambda x: calculate_mode(x),
    'danceability': 'mean',
    'energy': 'mean',
    'key': 'mean',
    'loudness': 'mean',
    'mode': 'mean',
    'speechiness': 'mean',
    'acousticness': 'mean',
    'instrumentalness': 'mean',
    'liveness': 'mean',
    'valence': 'mean',
    'tempo': 'mean',
    'duration_ms': 'mean',
    'time_signature': lambda x: calculate_mode(x),
    'Genre 1': lambda x: calculate_mode(x),
    'Genre 2': lambda x: calculate_mode(x),
    'Genre 3': lambda x: calculate_mode(x)
}

profile_df = df.groupby('userid').agg(aggregation_functions)

# Reordering columns to bring 'userid' to the front
columns = ['userid'] + [col for col in profile_df.columns if col != 'userid']
profile_df = profile_df[columns]

# Reset index if needed (depends on your preference if 'userid' should be an index or a column)
profile_df.reset_index(drop=True, inplace=True)

profile_df

#slice dataset into half to reduce the size of the dataset
df = df.iloc[:len(df)//2]
```
